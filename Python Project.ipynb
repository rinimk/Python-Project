```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Loading Given Dataset

data=pd.read_excel('myexcel.xlsx')
#data.info()
#data.describe()
#data.head(2)

#Preprocessing Height Column

data['Height']=data['Height'].apply(lambda x: np.random.randint(150,181) if isinstance(x,str) else x)

#Task 1: Distribution of employees across teams

teamcount = data['Team'].value_counts()
employeedistribution = (teamcount / len(data)) * 100

#Task1 Visualization

plt.figure(figsize=(8,6))
teamcount.plot(kind='bar',color='green')
plt.title('Distribution of Employees Across Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
    
```

```python
#Task2: Segregating employees based on their positions within the company. 

positions=data['Position'].value_counts()

# Task2 Visualization

plt.figure(figsize=(8,4))
positions.plot(kind='pie',autopct='%1.1f%%')
plt.title('Distribution of Employees by Position')
plt.ylabel('')
plt.show()
```

```python
 #Task3:Identify the predominant age group among employees

agegroups = pd.cut(data['Age'], bins=[20, 25, 30, 35, 40, 45, 50, 60, 70])
agegroupcounts = agegroups.value_counts()

# Task3 Visualization

plt.figure(figsize=(8,6))
agegroupcounts.plot(kind='bar', color='green')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
#Task4:Discovering the team and position with the highest salary expenditure.

salaryteam=data.groupby('Team')['Salary'].sum()

#Task4 Visualization

plt.figure(figsize=(8,6))
salaryteam.plot(kind='bar',color='green')
plt.title('Teamwise Salary Expenditure')
plt.xlabel('Team')
plt.ylabel('Salary Expenditure')
plt.xticks(rotation=45)
plt.show()
```

```python
#Task5:Correlation between Age and Salary 

correlation=data['Age'].corr(data['Salary'])

#Task5 Visualization

plt.figure(figsize=(8,6))
sns.scatterplot(x=data['Age'],y=data['Salary'])
plt.title('Age vs Salary Correlation')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()

```
